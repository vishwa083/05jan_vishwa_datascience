# Q1 Visualizing Chipotle's Data

```python
#Step 1. Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Step 2: Import the dataset from the provided address
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

```python
#Step 3. Assign it to a variable called chipo.
chipo = pd.read_csv(url, sep='\t')
```

```python
#Step 4. See the first 10 entries
print(chipo.head(10))
```

```python
#Step 5. Create a histogram of the top 5 items bought
top_items = chipo['item_name'].value_counts().head(5)
top_items.plot(kind='bar', title='Top 5 Items Bought')
plt.xlabel('Item Name')
plt.ylabel('Number of Orders')
plt.show()
```

```python
#Step 6. Create a scatterplot with the number of items orderered per order price
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))  
scatter_data = chipo.groupby('order_id').agg({'item_price': 'sum', 'quantity': 'sum'})
plt.scatter(scatter_data['item_price'], scatter_data['quantity'])
plt.title('Scatterplot of Items Ordered per Order Price')
plt.xlabel('Order Price')
plt.ylabel('Number of Items Ordered')
plt.show()
```

```python
#Step 7. BONUS: Create a question and a graph to answer your own question.
plt.hist(chipo.groupby('order_id')['item_price'].sum(), bins=20, edgecolor='black')
plt.title('Distribution of Order Prices')
plt.xlabel('Order Price')
plt.ylabel('Number of Orders')
plt.show()
```

# Q2 - Online Retails Purchase

```python
#Step 1. Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Step 2. Import the dataset from this address.
url ='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv'
```

```python
#Step 3. Assign it to a variable called online_rt
online_rt = pd.read_csv(url, encoding='latin1')
```

```python
#Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK
top_countries = online_rt[online_rt['Country'] != 'UK'].groupby('Country')['Quantity'].sum().nlargest(10)
top_countries.plot(kind='bar', title='Top 10 Countries by Quantity Ordered (excluding UK)')
plt.xlabel('Country')
plt.ylabel('Total Quantity Ordered')
plt.show()
```

```python
#Step 5. Exclude negative Quantity entries
online_rt = online_rt[online_rt['Quantity'] > 0]
```

```python
#Step 6. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries (except UK)
top3_countries = online_rt[online_rt['Country'].isin(['Netherlands', 'EIRE', 'Germany'])]
scatter_data = top3_countries.groupby(['CustomerID', 'Country']).agg({'Quantity': 'sum', 'UnitPrice': 'mean'})
plt.scatter(scatter_data['UnitPrice'], scatter_data['Quantity'])
plt.title('Scatterplot of Quantity per UnitPrice by CustomerID (Top 3 Countries)')
plt.xlabel('Average Unit Price')
plt.ylabel('Total Quantity Ordered')
plt.show()
```

```python
#Step 7. Investigate why the previous results look so uninformative.
#Step 7.1.1 Display the first few rows of that DataFrame
print(top3_countries.head())
```

```python
#Step 7.1.2 Think about what that piece of code does and display the dtype of UnitPrice
print(top3_countries['UnitPrice'].dtype)
```

```python
#Step 7.1.3 Pull data from online_rt for CustomerID s 12346.0 and 12347.0.
print(top3_countries[top3_countries['CustomerID'].isin([12346.0, 12347.0])])
```

```python
#Step 7.2 Reinterpreting the initial problem.
#Step 7.2.1 Find out the top 3 countries in terms of sales volume.
top3_countries_sales_volume = online_rt.groupby('Country')['Quantity'].sum().nlargest(3)
print(top3_countries_sales_volume)
```

```python
#Step 7.3 Modify, select and plot data
#Step 7.3.1 Add a column to online_rt called Revenue calculate the revenue (Quantity * UnitPrice) from each sale.
online_rt['Revenue'] = online_rt['Quantity'] * online_rt['UnitPrice']
```

```python
#Step 7.3.2 Group by CustomerID and Country and find out the average price ( AvgPrice ) each customer spends per unit.
avg_price_per_customer = online_rt.groupby(['CustomerID', 'Country']).agg({'Revenue': 'sum', 'Quantity': 'sum'})
avg_price_per_customer['AvgPrice'] = avg_price_per_customer['Revenue'] / avg_price_per_customer['Quantity']
```

```python
#Step 7.3.3 Plot
plt.scatter(avg_price_per_customer['AvgPrice'], avg_price_per_customer['Quantity'])
plt.title('Scatterplot of Quantity per AvgPrice by CustomerID')
plt.xlabel('Average Price per Unit')
plt.ylabel('Total Quantity Ordered')
plt.show()
```

```python
#Step 7.4 What to do now?
#Step 7.4.1 Plot the data for each CustomerID on a single graph
plt.scatter(avg_price_per_customer['AvgPrice'], avg_price_per_customer['Quantity'], c='blue', alpha=0.5, label='CustomerID')
plt.scatter(scatter_data['UnitPrice'], scatter_data['Quantity'], c='red', alpha=0.5, label='Top 3 Countries')
plt.title('Scatterplot of Quantity per AvgPrice by CustomerID')
plt.xlabel('Average Price per Unit')
plt.ylabel('Total Quantity Ordered')
plt.legend()
plt.show()
```

```python
#Step 7.4.2 Zoom in so we can see that curve more clearly
plt.scatter(avg_price_per_customer['AvgPrice'], avg_price_per_customer['Quantity'], c='blue', alpha=0.5, label='CustomerID')
plt.scatter(scatter_data['UnitPrice'], scatter_data['Quantity'], c='red', alpha=0.5, label='Top 3 Countries')
plt.title('Scatterplot of Quantity per AvgPrice by CustomerID (Zoomed In)')
plt.xlabel('Average Price per Unit')
plt.ylabel('Total Quantity Ordered')
plt.legend()
plt.xlim(0, 50)  # Adjust the x-axis range for better visibility
plt.show()
```

```python
#8. Plot a line chart showing revenue (y) per UnitPrice (x)
revenue_per_price = online_rt.groupby('UnitPrice')['Revenue'].sum()
plt.plot(revenue_per_price.index, revenue_per_price)
plt.title('Revenue per UnitPrice')
plt.xlabel('Unit Price')
plt.ylabel('Total Revenue')
plt.show()
```

```python
#8.1 Group UnitPrice by intervals of 1 for prices [0,50), and sum Quantity and Revenue .
price_intervals = pd.cut(online_rt['UnitPrice'], bins=range(0, 51, 1), right=False)
grouped_data = online_rt.groupby(price_intervals).agg({'Quantity': 'sum', 'Revenue': 'sum'})
```

```python
# 8.3 Plot.
# Convert the midpoints of the intervals to float
midpoints = grouped_data.index.categories.mid

# Plot the line chart
plt.plot(midpoints, grouped_data['Revenue'])
plt.title('Revenue per UnitPrice (Interval of 1)')
plt.xlabel('Unit Price')
plt.ylabel('Total Revenue')
plt.xticks(range(0, 51, 5))  # Set x-axis ticks for better readability
plt.show()
```

# Q3 - Scores

```python
#Step 1. Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Step 2. Create the DataFrame that should look like the one below.
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'age': [42, 52, 36, 24, 73],
    'female': [0, 1, 1, 0, 1],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age'], alpha=0.7)
plt.title('Scatterplot of preTestScore and postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

```python
#Step 4. Create a Scatterplot of preTestScore and postTestScore.
colors = df['female'].map({0: 'blue', 1: 'pink'})
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['postTestScore'] * 4.5, c=colors, alpha=0.7)
plt.title('Scatterplot of preTestScore and postTestScore (Size and Color by sex)')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

```python
#BONUS: Create your own question and answer it.
#Example: What is the correlation between age and postTestScore?
correlation = df['age'].corr(df['postTestScore'])
print(f'Correlation between age and postTestScore: {correlation}')
```

# Q4 - Tips

```python
#Step 1. Import the necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Step 2. Import the dataset from this address.
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
```

```python
#Step 3. Assign it to a variable called tips
tips = sns.load_dataset('tips')
```

```python
#Step 4. Delete the Unnamed 0 column
tips = tips.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
#Step 5. Plot the total_bill column histogram
sns.histplot(tips['total_bill'], bins=30, kde=True)
plt.title('Histogram of total_bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
#Step 6. Create a scatter plot presenting the relationship between total_bill and tip
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter plot of total_bill and tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
#Step 7. Create one image with the relationship of total_bill, tip and size.
sns.pairplot(tips, vars=['total_bill', 'tip', 'size'])
plt.show()
```

```python
#Step 8. Present the relationship between days and total_bill value
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Relationship between days and total_bill')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
#Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex
sns.scatterplot(x='tip', y='day', hue='sex', data=tips)
plt.title('Scatter plot of tip and day (differentiated by sex)')
plt.xlabel('Tip')
plt.ylabel('Day')
plt.show()
```

```python
#Step 10. Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
plt.title('Box plot of total_bill per day (differentiated by time)')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
#Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.
sns.histplot(x='tip', hue='time', multiple='stack', data=tips)
plt.title('Histograms of tip based on Dinner and Lunch')
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.show()
```

```python
#Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill valuue and tip relationship, differing by smoker or no smoker
sns.relplot(x='total_bill', y='tip', hue='smoker', col='sex', data=tips)
plt.show()
```

```python
#BONUS:Create your own question and answer it using a graph
# Example: What is the average tip amount for different days?
sns.barplot(x='day', y='tip', data=tips)
plt.title('Average Tip Amount for Different Days')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.show()
```

# Q5 - Visualizing the Titanic Disaster

```python
#Step 1. Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Step 2. Import the dataset from this address
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
```

```python
#Step 3. Assign it to a variable titanic
titanic = pd.read_csv(url)
```

```python
#Step 4. Set PassengerId as the index
titanic.set_index('PassengerId', inplace=True)
```

```python
#Step 5. Create a pie chart presenting the male/female proportion
gender_counts = titanic['Sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Male/Female Proportion on Titanic')
plt.show()
```

```python
#Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender
sns.scatterplot(x='Fare', y='Age', hue='Sex', data=titanic)
plt.title('Scatterplot of Fare and Age (Color by Gender)')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.show()
```

```python
#Step 7. How many people survived?
survived_count = titanic['Survived'].sum()
print(f'Total number of people who survived: {survived_count}')
```

```python
#Step 8. Create a histogram with the Fare payed
sns.histplot(titanic['Fare'], bins=30, kde=True)
plt.title('Histogram of Fare Paid on Titanic')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
#BONUS: Create your own question and answer it.
# Example: What is the survival rate for different classes?
sns.barplot(x='Pclass', y='Survived', data=titanic)
plt.title('Survival Rate for Different Classes')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```
