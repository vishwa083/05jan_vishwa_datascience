```python
#A. Creating and customizing line charts using Matplotlib
iris = sns.load_dataset('iris')    #load iris dataset
iris

selected_feature = 'sepal_length'   #selected a specific feature

for species in iris['species'].unique():                       #created a line chart
    species_data = iris[iris['species'] == species]
    plt.plot(species_data[selected_feature], label=species)

plt.title('Species')        #customizing the plot
plt.xlabel('Data Point Index')
plt.ylabel(selected_feature)
plt.legend(loc='upper left')
plt.show()
```

```python
#B. Visualizing relationships between two or more variables using scatter plots
iris = sns.load_dataset('iris')             #load iris dataset
sns.pairplot(iris, hue='species')
plt.suptitle('Pairplot')
plt.show()
#pairplot is use to create a grid of scatterplots which shows relation between pairs of variables
```

```python
#C. Studying distributions of variables using histograms & bar charts to
iris = sns.load_dataset('iris')  

mean_data = iris.groupby('species').mean()      #calculate mean values

mean_data.plot(kind='bar')       # bar plot
plt.title('Features')
plt.xlabel('species')
plt.ylabel('Mean Values')
plt.legend(loc='upper left')
plt.show()
```

```python
#D. Visualizing two-dimensional data using heatmaps
iris = sns.load_dataset('iris')   
columns = iris.select_dtypes(include=['float64', 'int64'])  #filter and select only the columns in iris
matrix = columns.corr()
plt.figure(figsize=(10, 8))                        
sns.heatmap(matrix, annot=True, cmap='coolwarm')   
plt.title('Heatmap')                               
plt.show()

# cmap:set the color map for the heatmap
# annot=true:display numeric value in each cell
```

```python
#F. Plotting multiple Matplotlib and Seaborn charts in a grid
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x
y4 = x**2
fig, axes = plt.subplots(2, 2)            # 2x2 grid for subplots
axes[0, 0].plot(x, y1)                    # line plot
axes[0, 1].scatter(x, y2, color='red')                 # scatter plot
axes[1, 0].hist(y3, color='green', alpha=0.7)                       # histogram
axes[1, 1].bar(x, y4, color='purple', alpha=0.7)                     # bar plot
plt.show()
```

```python
#10. Create an array of numbers between 0 and 6 with increments of 0.3 and name its â€œxâ€. Then on the same plot, plot x, xÂ², xÂ³, and xâ´. For consistency, use the following style lines respectively, â€œroâ€, â€œbsâ€, â€œgâ€, and â€œ:â€. Lastly, make sure that the x-axis covers 0 to 6, whilethe y-axis spans from 0 to 125. Do not worry if you are familiar with the style lines â€” you will recognize them as soon as you see the plot.
import numpy as np
x = np.arange(0, 6.3, 0.3)
plt.plot(x, x, 'ro', label='x')
plt.plot(x, x**2, 'bs', label='xÂ²')
plt.plot(x, x**3, 'g', label='xÂ³')
plt.plot(x, x**4, ':', label='xâ´')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([0, 6, 0, 125])
plt.legend()
plt.show()
```
